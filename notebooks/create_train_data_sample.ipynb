```python
import pandas as pd
from sklearn.datasets import make_classification
```

```python
n_features = 11

X, y = make_classification(n_samples=10**4, n_features=n_features, random_state=42)
X.shape, y.shape
```

```python
df = pd.DataFrame({
    f'col_{i}': X[:, i]
    for i in range(n_features)
})
df['target'] = y
df.head()
```

```python
# make some columns categorial
import numpy as np

def to_cat(x: pd.Series, n_cat=5):
    l = np.linspace(x.min(), x.max(), n_cat+1)
    return x.apply(lambda el: np.argmax(el < l))

for i in range(5, 11):
    df[f'col_{i}'] = to_cat(df[f'col_{i}'], n_cat=i)

df.head()
```

```python
df.to_csv('../tests/train_data_sample.csv', index=0)
```
